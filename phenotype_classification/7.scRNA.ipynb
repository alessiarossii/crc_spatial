## Import

```python
import os
import scanpy as sc
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
path = "output_phenotypes"
```

## Load

```python
# single cell data
sc_adata = sc.read_h5ad("../../../../../datasets/Marteau_2024_CRC/final_crc_atlas-adata.h5ad")
```

```python
# dataframe with the phenotype assignments to each patient
df = pd.read_csv('../input data/scRNAseq.tsv', sep='\t')
```

## Extract MUI data

```python
adata = sc_adata[sc_adata.obs['dataset'] == 'MUI_Innsbruck']
adata = adata[adata.obs['sample_type'] != 'blood'].copy()
```

```python
adata.obs['tumor_source'].unique() # FOR THE SC WE DON'T HAVE MARGIN SAMPLES ???
```

## Map phenotypes on sc data

```python
# extract from 'patient_id' the final part 'P_'
adata.obs['patient_short'] = adata.obs['patient_id'].str.split('.').str[-1]

# dict for mapping
mapping = df.set_index('scRNA-seq')['Type'].to_dict()

# mapping of phenotype
adata.obs['phenotype'] = adata.obs['patient_short'].map(mapping)
```

```python
adata.obs['patient_short'].unique().tolist()
```

```python
adata = adata[adata.obs['patient_short'] != 'P15'].copy()
```

## Save

```python
adata.write(os.path.join(path, 'sc_phenotypes.h5ad'))
```
